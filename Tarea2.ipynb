$$
\sum_{i=1}^{n} p_{i}\left(\left|a_{i}^{3}\right|-\alpha\left|a_{i}\right|\right) \leq 0
$$
(d) La restricción es equivalente a una desigualdad lineal
$$
\sum_{i=1}^{n} p_{i} a_{i}^{2} \leq \alpha
$$
(e) La restricción es equivalente a una desigualdad lineal
$$
\sum_{i=1}^{n} p_{i} a_{i}^{2} \geq \alpha .
$$
Por lo tanto, las primeras cinco restricciones definen conjuntos convexos.

(f) La restricción
$$
\operatorname{var}(x)=\mathbf{E} x^{2}-(\mathbf{E} x)^{2}=\sum_{i=1}^{n} p_{i} a_{ i}^{2}-\left(\sum_{i=1}^{n} p_{i} a_{i}\right)^{2} \leq \alpha
$$
no es convexo en general. Como contraejemplo, podemos tomar $n=2, a_{1}=0, a_{2}=1$ y $\alpha=1 / 5 . p=(1,0)$ y $p=(0,1)$ son dos puntos que satisfacen $\operatorname{var}(x) \leq \alpha$, pero la combinación convexa $p=(1 / 2, 1 / 2)$ no lo hace.

(g) Esta restricción es equivalente a
$$
\sum_{i=1}^{n} a_{i}^{2} p_{i}+\left(\sum_{i=1}^{n} a_{i} p_{i}\right)^ {2}=b^{T} p+p^{T} A p \leq \alpha
$$
donde $b_{i}=a_{i}^{2}$ y $A=a a^{T}$. Esto define un conjunto convexo, ya que la matriz $a a^{T}$ es semidefinida positiva.

Denotemos el cuartil $(x)=f(p)$ para enfatizar que es una función de $p$. La figura ilustra la definición. Muestra la distribución acumulada para una distribución $p$ con $f(p)=a_{2}$.

(h) La restricción $f(p) \geq \alpha$ es equivalente a
$$
\operatorname{prob}(x \leq \beta)<0.25 \text { para todos } \beta<\alpha
$$

Esta es una desigualdad lineal estricta en $p$, que define un semiespacio abierto.

\operatorname{prob}(x \leq \beta) \geq 0.25 \text { para todos } \beta \geq \alpha
$$
Esto se puede expresar como una desigualdad lineal
$$
\sum_{i=k+1}^{n} p_{i} \geq 0.25 .

$2.17$ Image of polyhedral sets under perspective function. In this problem we study the image of hyperplanes, halfspaces, and polyhedra under the perspective function $P(x, t)=x / t$, with $\operatorname{dom} P=\mathbf{R}^{n} \times \mathbf{R}_{++}$. For each of the following sets $C$, give a simple description of
$$
P(C)=\{v / t \mid(v, t) \in C, t>0\} .
$$

(a) The polyhedron $C=\operatorname{conv}\left\{\left(v_{1}, t_{1}\right), \ldots,\left(v_{K}, t_{K}\right)\right\}$ where $v_{i} \in \mathbf{R}^{n}$ and $t_{i}>0$.

(b) The hyperplane $C=\left\{(v, t) \mid f^{T} v+g t=h\right\}$ (with $f$ and $g$ not both zero).

(c) The halfspace $C=\left\{(v, t) \mid f^{T} v+g t \leq h\right\}$ (with $f$ and $g$ not both zero).

(d) The polyhedron $C=\{(v, t) \mid F v+g t \preceq h\}$.

Solución:

(a) El poliedro $C=\operatorname{conv}\left\{\left(v_{1}, t_{1}\right), \ldots,\left(v_{K}, t_{K}\right )\right\}$ donde $v_{i} \in \mathbf{R}^{n}$ y $t_{i}>0$. Solución. el poliedro
$$
P(C)= conv\left\{v_{1} / t_{1}, \ldots, v_{K} / t_{K}\right\} .
$$
Primero mostramos que $P(C) \subseteq \operatorname{conv}\left\{v_{1} / t_{1}, \ldots, v_{K} / t_{K}\right\}$. Sea $x=(v, t) \in C$, con
$$
v=\sum_{i=1}^{K} \theta_{i} v_{i}, \quad t=\sum_{i=1}^{K} \theta_{i} t_{i},
$$
y $\theta \succeq 0, \mathbf{1}^{T} \theta=1$. La imagen $P(x)$ se puede expresar como
$$
P(x)=v / t=\frac{\sum_{i=1}^{K} \theta_{i} v_{i}}{\sum_{i=1}^{K} \theta_{i} t_{i}}=\sum_{i=1}^{K} \mu_{i} v_{i} / t_{i}
$$
donde
$$
\mu_{i}=\frac{\theta_{i} t_{i}}{\sum_{k=1}^{K} \theta_{k} t_{k}}, \quad i=1, \ldots , K .
$$
Está claro que $\mu \succeq 0, \mathbf{1}^{T} \mu=1$, por lo que podemos concluir que $P(x) \in \operatorname{conv}\left\{v_{1 } / t_{1}, \ldots, v_{K} / t_{K}\right\}$ para todo $x \in C$.
A continuación, mostramos que $P(C) \supseteq \operatorname{conv}\left\{v_{1} / t_{1}, \ldots, v_{K} / t_{K}\right\}$. Considere un punto
$$
z=\sum_{i=1}^{K} \mu_{i} v_{i} / t_{i}
$$
con $\mu \succeq 0, \mathbf{1}^{T} \mu=1$. Definir
$$
\theta_{i}=\frac{\mu_{i}}{t_{i} \sum_{j=1}^{K} \mu_{j} / t_{j}}, \quad i=1, \ Idots, K.
$$
Está claro que $\theta \succeq 0$ y $\mathbf{1}^{T} \theta=1$. Además, $z=P(v, t)$ donde
$$
t=\sum_{i} \theta_{i} t_{i}=\frac{\sum_{i} \mu_{i}}{\sum_{j} \mu_{j} / t_{j}}=\ frac{1}{\sum_{j} \mu_{j} / t_{j}}, \quad v=\sum_{i} \theta_{i} v_{i}
$$
es decir, $(v, t) \in C$.

(b) el hiperplano $C=\left\{(v, t) \mid f^{T} v+g t=h\right\}$ (con$f$ y $g$ ambos diferentes de cero).
$$
\begin{aligned}
P(C) &=\left\{z \mid f^{T} z+g=h / t \text { for some } t>0\right\} \\
&=\left\{\begin{array}{ll}
\left\{z \mid f^{T} z+g=0\right\} & h=0 \\
\left\{z \mid f^{T} z+g>0\right\} & h>0 \\
\left\{z \mid f^{T} z+g<0\right\} & h<0 .
\end{array}\right.
\end{aligned}
$$
(c) el semiespacio $C=\left\{(v, t) \mid f^{T} v+g t \leq h\right\}$ (con $f$ y $g$ diferentes de cero).
$$
\begin{aligned}
P(C) &=\left\{z \mid f^{T} z+g \leq h / t \text { for some } t>0\right\} \\
&=\left\{\begin{array}{ll}
\left\{z \mid f^{T} z+g \leq 0\right\} & h=0 \\
\mathbf{R}^{n} & h>0 \\
\left\{z \mid f^{T} z+g<0\right\} & h<0 .
\end{array}\right.
\end{aligned}
$$
(d) el poliedro $C=\{(v, t) \mid F v+g t \preceq h\}$.
$$
P(C)=\{z \mid F z+g \preceq(1 / t) h \text { for some } t>0\} .
$$

3. Verificar que el conjunto de soluciones de un conjunto de desigualdades de matrices lineales es convexa.

sea el  cojunto de desigualdades de matrices lineales definido como:
$$
D=\{x \mid A x \preceq b\}
$$
sea $x, y \in D$ veamos que $\theta x+(1-\theta) y \in D$ con $0 \leq \theta \leq 1$. Multiplicando por la matriz A tenemos:
$$
A(\theta x+(1-\theta) y)=A \theta x+A(1-\theta) y=\theta A x+(1-\theta) A y
$$
como $x . y \in A$ entonces vamos a tener que $A x \preceq b$ y $A y \preceq b$, con esto vamos a tener que:
$$
\theta A x+(1-\theta) A y \preceq \theta b+(1-\theta) b=b
$$
Por lo tanto como $\theta x+(1-\theta) y \in D$ entonces D es convexo
